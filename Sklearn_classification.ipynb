```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('classification.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
data['diagnosis'] = data['diagnosis'].map({'M':1,'B':0})
```

```python
data.head()
```

```python
f,ax = plt.subplots(figsize=(18, 18))
sns.heatmap(data.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
```

```python
data = data.drop(['id','Unnamed: 32','texture_mean','smoothness_mean','symmetry_mean','fractal_dimension_mean','texture_se','smoothness_se',
                  'compactness_se','concavity_se','concave points_se','symmetry_se','fractal_dimension_se','smoothness_worst','symmetry_worst',
                 'fractal_dimension_worst'],axis = 1)
```

```python
x = data.iloc[:, 1:-1]
y = data.iloc[:, 0]
```

```python
# Create train and test datasets
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42, stratify = y)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
y_pred = logreg.predict(x_test)
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred )
print(cm)
acclr_ = accuracy_score(y_test, y_pred)
```

```python
# Naive Bayes Theorem
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(x_train, y_train) 
y_pred = classifier.predict(x_test)
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm_nb = confusion_matrix(y_test, y_pred )
print(cm_nb)
accnb_ = accuracy_score(y_test, y_pred)
```

```python
print('Accuracy Score of Logistic Regression is :',acclr_,'\nAccuracy Score of Naive Bayes is :',accnb_)
```
