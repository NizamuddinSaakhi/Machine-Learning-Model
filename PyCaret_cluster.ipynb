## Clustering Module
PyCaret's Clustering module is an unsupervised machine learning module that performs the task of gathering a set of objects in such a way that object within the same group(moreover known as Cluster) are more comparable to each other than to those in other group.

## Import Libraries

```python
import pandas as pd
```

### Read Dataset
The dataset is about mall customer, each rows correspond to each customer. and each customer has CustomerID, Gerne, Age, Annual Income(k$) and Spending Score(1-100). Here Spending Score has value from 1 to 100 which basically mean those customer has nearest number to 100 that customer spend more money and those customer has nearest to 1 that customer spend less money for shopping in the mall. 

```python
data = pd.read_csv('Cluster.csv')
data.head()
```

### Setting up Environment
This function initialize the environment in PyCaret. setup() must called before executing any other work in pycaret. it takes only one mandatory parameter which is data. here i have ignore some feature for better understanding.

```python
from pycaret.clustering import *
clas = setup(data = data, ignore_features=['Genre', 'Age', 'CustomerID'])
```

## Creating Model
Create model will train the model, PyCaret offer's number of clustering model such as Mean Shift Clustering, K-modes Clustering and so on. but i have use one of the usefull clustering method which is k-means clustering.

```python
crt = create_model('kmeans')
```

### Visualising Prediction Result

### Plot Graph for the clustering
plot parameter take trained dataset and return the visualized model which I have selected, Here i have selected Silhouette graph for better understading of our cluster output.

```python
 plot = plot_model(crt, plot='silhouette')
```

### Evaluate Model
Evaluate parameter take trained data and return the visualized the model but here Evaluate model will give options which model do you want to create. Here i have selected two graph which are Cluster PCA Plot in (2D) and another graph is Elbow
